```python
from bs4 import BeautifulSoup
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
from selenium import webdriver
import time
from selenium.webdriver.common.by import By
import re
```

```python
driver = webdriver.Chrome('/home/huudu/y4/dp/spotify-to-mp3-python/chromedriver')
```

```python
driver.get('https://open.spotify.com/')
```

```python
df = pd.read_csv('/home/huudu/y4/dp/spotify-to-mp3-python/Gin Tuấn Kiệt Radio/Gin Tuấn Kiệt Radio.txt')

for i in range(df.shape[0]):
    url = df.iloc[i][2]
    title = df.iloc[i][0]
    driver.get(url)
    time.sleep(3)
    lyrics = []
    try:
        content  = driver.find_element(By.CSS_SELECTOR,'.Q3OKWaFrTVTIRZyG05Gh')
        content = content.find_elements(By.TAG_NAME,'p')
        for p in content:
            txt = p.get_attribute('innerHTML')
            # txt = re.sub(r'[^\w\s]','',txt)
            txt = txt.replace(',','')
            lyrics.append(txt)
        pd.DataFrame([{'title':title,'lyric':" ".join(lyrics)}]).to_csv('lyrics.csv',mode='a', header=False, index=False, encoding='utf-8-sig')
    except Exception as e:
        print(e)
```
